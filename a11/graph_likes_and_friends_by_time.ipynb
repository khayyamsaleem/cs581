# Graphing change in friends added and like counts over time on FB

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (15,10)
```

```python
data = pd.read_json("combined.json")
data = data.reindex(sorted(data.columns), axis=1)
```

```python
dates = list(data.columns.values)
```

```python
num_friends_added_by_month = list(data.iloc[0])
react_counts_by_month = list(data.iloc[1])
```

```python
plt.plot(dates, num_friends_added_by_month, 'ro', label="# friends added")
plt.plot(dates, react_counts_by_month, 'bo', label="# of reacts")
plt.legend(loc=2)
plt.show()
```

## Summary

After initially downloading my personal data from facebook's personal data dump tool, I decided to see if there were any trends in the number of friends I added per month, and the total number of "reacts" I would give out to posts each month (interactions with FB content or comments in a given month). What I was able to notice was that rather than having correlations between each other, they correlated more closely with events or changes in my life. 

Based on the data retrieval, formatting, the above visualization and the ensuing analysis, I was able to identify that I would have in uptick in number of friend requests at the beginning of every new school year, starting in high school. I would also "like" or "react" to lots of content more often towards big events, such as my high school graduation or beginning of first semester of university. 

Notably, my activity in additional friend requests slows over summers, while my number of reactions to posts and comments increases over summers. Contrarily, my number of friend requests increases at the beginning of school years but my number of reacts typically decreases. This is correspondent with the amount of free time I typically have available at these times of year, as well as the times of year during which I meet the most new people.
